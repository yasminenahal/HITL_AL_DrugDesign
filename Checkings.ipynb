```python
def uncertainty_sampling_entropy(prob_dists):
    prbslogs = prob_dists * np.log2(prob_dists)
    numerator = 0 - np.sum(prbslogs)
    denominator = np.log2(prob_dists.size)
    entropy = numerator / denominator
    return entropy
```

```python
def plot_trace(param, param_name='parameter'):
    """Plot the trace and posterior of a parameter."""
  
    # Summary statistics
    mean = np.mean(param)
    median = np.median(param)
    cred_min, cred_max = np.percentile(param, 2.5), np.percentile(param, 97.5)
  
    # Plotting
    plt.subplot(2,1,1)
    plt.plot(param)
    plt.xlabel('samples')
    plt.ylabel(param_name)
    plt.axhline(mean, color='r', lw=2, linestyle='--')
    plt.axhline(median, color='c', lw=2, linestyle='--')
    plt.axhline(cred_min, linestyle=':', color='k', alpha=0.2)
    plt.axhline(cred_max, linestyle=':', color='k', alpha=0.2)
    plt.title('Trace and Posterior Distribution for {}'.format(param_name))

    plt.subplot(2,1,2)
    plt.hist(param, 30, density=True); sns.kdeplot(param, fill=True)
    plt.xlabel(param_name)
    plt.ylabel('density')
    plt.axvline(mean, color='r', lw=2, linestyle='--',label='mean')
    plt.axvline(median, color='c', lw=2, linestyle='--',label='median')
    plt.axvline(cred_min, linestyle=':', color='k', alpha=0.2, label='95% CI')
    plt.axvline(cred_max, linestyle=':', color='k', alpha=0.2)
  
    plt.gcf().tight_layout()
    plt.legend()
```

```python
import pickle
drd2_oracle = pickle.load(open("/home/klgx638/Projects/reinvent-hitl-calibration/clf_p3.pkl", "rb"))

def report_results_no_human(scaff, setting = "regression", objective = "logp", feedback_step = 100, min_high_score = 0.8, max_mols = 2000):
    
    best_mols_pred = []
    best_mols_true = []
    j1 = 0
    j2 = feedback_step
    
    for i in range(10):
        last_iter = scaff[scaff["Step"] > j1]
        last_iter = scaff[scaff["Step"] < j2]
        best_last_iter = last_iter[last_iter["bioactivity"] > min_high_score].sort_values(by = "bioactivity")
        best_smiles_last_iter = best_last_iter.SMILES
        best_mols = [Chem.MolFromSmiles(s) for s in best_smiles_last_iter]
        if objective == "logp":
            if setting == "regression":
                best_mols_pred.append(best_last_iter.raw_bioactivity.values.tolist())
                best_mols_true.append([MolLogP(m) for m in best_mols])
            if setting == "classification":
                best_mols_pred_scores = best_last_iter.bioactivity.values.tolist()
                best_mols_pred.append([1 if s > 0.5 else 0 for s in best_mols_pred_scores])
                best_mols_true.append([1 if MolLogP(m) > 1 and MolLogP(m) < 2 else 0 for m in best_mols])
        if objective == "drd2":
            best_mols_pred_scores = best_last_iter.bioactivity.values.tolist()
            best_mols_pred.append([1 if s > 0.5 else 0 for s in best_mols_pred_scores])
            if len(best_mols) >= 1: 
                best_mols_true.append(drd2_oracle.predict(fingerprints_from_mol(best_mols)).tolist())
            else:
                best_mols_true.append([])
        j1 += feedback_step
        j2 += feedback_step
    return best_mols_pred, best_mols_true

def report_results_w_human(scaffs, objective = "logp", setting = "regression", feedback_step = 100, min_high_score = 0.8, max_mols = 2000):  
    best_mols_pred = []
    best_mols_true = []
    j1 = 0
    j2 = feedback_step
    for i in range(10):
        last_iter = scaffs[i]
        best_last_iter = last_iter[last_iter["bioactivity"] > min_high_score].sort_values(by = "bioactivity")
        best_smiles_last_iter = best_last_iter.SMILES
        best_mols = [Chem.MolFromSmiles(s) for s in best_smiles_last_iter]
        if objective == "logp":
            if setting == "regression":
                best_mols_pred.append(best_last_iter.raw_bioactivity.values)
                best_mols_true.append([MolLogP(m) for m in best_mols])
            if setting == "classification":
                best_mols_pred_scores = best_last_iter.bioactivity.values.tolist()
                best_mols_pred.append([1 if s > 0.5 else 0 for s in best_mols_pred_scores])
                best_mols_true.append([1 if MolLogP(m) > 1 and MolLogP(m) < 2 else 0 for m in best_mols])
        if objective == "drd2":
            best_mols_pred_scores = best_last_iter.bioactivity.values.tolist()
            best_mols_pred.append([1 if s > 0.5 else 0 for s in best_mols_pred_scores])
            if len(best_mols) >= 1: 
                best_mols_true.append(drd2_oracle.predict(fingerprints_from_mol(best_mols)).tolist())
            else:
                best_mols_true.append([])
            
        j1 += feedback_step
        j2 += feedback_step

    return best_mols_pred, best_mols_true
```
